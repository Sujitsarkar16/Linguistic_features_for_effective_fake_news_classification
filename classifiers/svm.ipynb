```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
import pandas as pd
import numpy as np
import pickle
import os
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv("../data/Significant_features.csv")
```

```python
data
```

```python
X = data.drop('label', axis=1)
y = data['label']
```

```python
y_encoded = pd.get_dummies(y, drop_first=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
svm_linear_model = SVC(kernel='linear') 
svm_linear_model.fit(X_train_scaled, y_train.values.ravel())
```

```python
y_pred_linear = svm_linear_model.predict(X_test_scaled)
accuracy_linear = accuracy_score(y_test, y_pred_linear)
```

```python
svm_rbf_model = SVC(kernel='rbf')
svm_rbf_model.fit(X_train_scaled, y_train.values.ravel())
```

```python
y_pred_rbf = svm_rbf_model.predict(X_test_scaled)
accuracy_rbf = accuracy_score(y_test, y_pred_rbf)
```

```python
accuracy_linear, accuracy_rbf
```

```python
classifier_location = "../src/fnClassification/Models/"
os.makedirs(classifier_location, exist_ok=True)
model_filename = os.path.join(classifier_location, 'SVM.pkl')
```

```python
with open(model_filename, 'wb') as file:
    pickle.dump(accuracy_linear, file)
```
