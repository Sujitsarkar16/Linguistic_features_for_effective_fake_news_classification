```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
import os
import pickle
```

```python
df = pd.read_csv("../data/cleaned_news.csv")
```

```python
df
```

```python
df = df.drop(["text"], axis=1)
df = df.drop(["date"], axis=1)
df = df.drop(["subject"], axis=1)
```

```python
df
```

```python
vectorizer = CountVectorizer(max_features=1000, ngram_range=(1,1), stop_words='english')
```

```python
vectorizer.fit(df["cleaned_text"])
```

```python
vectorized_data = vectorizer.transform(df["cleaned_text"])
```

```python
print("Feature names:", vectorizer.get_feature_names_out()[:40])  
print("Shape of the matrix:", vectorized_data.shape)
```

```python
classifier_location = "../src/fnClassification/Models/"
os.makedirs(classifier_location, exist_ok=True)  
model_filename = os.path.join(classifier_location, 'cvec.pkl')

with open(model_filename, 'wb') as file:
    pickle.dump(vectorizer, file)

print(f"CVEC model saved to: {model_filename}")
```
